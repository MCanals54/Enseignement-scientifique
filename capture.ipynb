### Problèmatique.
Un étang habrite une population de poissons dont on veut estimer le nombre $N$. Comme il est impossible de recenser tous les individus de l'étang on utilise une méthode appelée *capture - marquage - recapture*.

Cette méthode consiste à capturer un échantillon de taille $N_0$ de cette population et de marquer les individus. 

```python
import matplotlib.pyplot as plt
```

```python
from random import randint, choice
```

```python
size = randint(1000,2000)
```

```python
population = [ (randint(0,100), randint(0,100))  for _ in range(size)] 

```
